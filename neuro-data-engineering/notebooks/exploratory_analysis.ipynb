# Exploratory Data Analysis on ADHD, Autism, and Other Neurological Diseases

This notebook is dedicated to performing exploratory data analysis (EDA) on datasets related to ADHD, autism, and other neurological diseases. The goal is to uncover patterns, relationships, and insights that can inform further analysis and research.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style='whitegrid')
```

```python
# Load the processed data
processed_data_path = '../data/processed/processed_data.csv'
data = pd.read_csv(processed_data_path)

# Display the first few rows of the dataset
data.head()
```

```python
# Summary statistics of the dataset
data.describe()
```

```python
# Visualize the distribution of a key variable
plt.figure(figsize=(10, 6))
sns.histplot(data['key_variable'], bins=30, kde=True)
plt.title('Distribution of Key Variable')
plt.xlabel('Key Variable')
plt.ylabel('Frequency')
plt.show()
```

```python
# Explore relationships between variables
plt.figure(figsize=(12, 8))
sns.scatterplot(x='variable_x', y='variable_y', data=data, hue='category_variable')
plt.title('Relationship between Variable X and Variable Y')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.legend(title='Category')
plt.show()
```
